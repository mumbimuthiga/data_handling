```python
import pandas as pd
titanic=pd.read_csv("/content/Titanic.csv")
titanic.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#print(titanic.columns)  # Debugging step
#sns.countplot(x="Survived",data=titanic,palette="coolwarm") #Creates a bar plot showing the count of each unique value in the survived column
sns.countplot(x="Survived",data=titanic ,palette="muted")
plt.title("Survival Count")
plt.xlabel("Survived(0=No,1=Yes)")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns;
import matplotlib.pyplot as plt
import warnings as warning;

warning.filterwarnings('ignore');
# Set visualization style
sns.set_style("whitegrid")

# Countplot for Survived column
plt.figure(figsize=(6, 4))
sns.countplot(x="Survived", data=titanic, palette="coolwarm")
plt.title("Survival Count")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()


plt.figure(figsize=(8,5))
sns.histplot(titanic["Age"].dropna(),bins=30,kde=True,palette="coolwarm")
#sns.histplot(df["Age"].dropna(), bins=30, kde=True, color="blue")
plt.title("Age Distribution")
plt.xlabel("Age Count")
plt.ylabel("Counts")
plt.show()

plt.figure(figsize=(6,4))
sns.barplot(x="Pclass",y="Survived", data=titanic ,palette="viridis")
plt.title("Survived vs PClass")
plt.xlabel("Pclass")
plt.ylabel("Survived")
plt.show()



plt.figure(figsize=(6,5))
sns.barplot(x="Sex",y="Survived",data=titanic,palette="coolwarm")
plt.title("Gender Vs Survived")
plt.xlabel("Gender")
plt.ylabel("Survived")
plt.show()
```

```python
titanicdf=titanic.copy()
titanicdf.drop(columns=[])
titanicdfcleaned=titanicdf.drop(columns=['Cabin' ,'Ticket'])# Drop columns with excessive missing values and less relevance
titanicdfcleaned.head()
# Fill missing values
titanicdfcleaned['Age'].fillna(titanicdfcleaned['Age'].median(),inplace=True) #using median
titanicdfcleaned['Embarked'].fillna(titanicdfcleaned['Embarked'].mode()[0],inplace=True) #Using mode

# Verify changes
titanicdfcleaned.info(), titanicdfcleaned.isnull().sum()
```

```python
titanicdfcleaned.head()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(6,4))
sns.countplot(x="Survived",data=titanicdfcleaned ,palette="coolwarm")
plt.title("Survived Distribution after cleaning");
plt.xlabel("Survived 1=Yes,0=No")
plt.ylabel("Count")
plt.show()



plt.figure(figsize=(8,5))
sns.histplot(x="Age",data=titanicdfcleaned,kde=True,bins=30 ,palette="coolwarm")
plt.title("Age Distribution after cleaning")
plt.xlabel("Age Distribution")
plt.ylabel("Count")
plt.show()
```
