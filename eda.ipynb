```python
import pandas as pd
titanic=pd.read_csv("/content/Titanic.csv")
titanic.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#print(titanic.columns)  # Debugging step
sns.countplot(x="Survived",data=titanic,palette="coolwarm") #Creates a bar plot showing the count of each unique value in the survived column

plt.title("Survival Count")
plt.xlabel("Survived(0=No,1=Yes)")
plt.ylabel("Count")
plt.show()
```
