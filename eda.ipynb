```python
import pandas as pd
titanic=pd.read_csv("/content/Titanic.csv")
titanic.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
#print(titanic.columns)  # Debugging step
#sns.countplot(x="Survived",data=titanic,palette="coolwarm") #Creates a bar plot showing the count of each unique value in the survived column
sns.countplot(x="Survived",data=titanic ,palette="muted")
plt.title("Survival Count")
plt.xlabel("Survived(0=No,1=Yes)")
plt.ylabel("Count")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(titanic['Age'].dropna(),bins=30,kde=False,color="blue")
#sns.histplot(titanic['Age'].dropna(),bins=50,kde=False,color="green")
#create a histogram with the column age
#dropna() drop missing values in the column
#let the data be distributed/split in groups of 30
#color is blue
#kde  adds a smooth probability density curve over the histogram.(Kernel Density Estimate)
plt.title("Age Distribution")
```
